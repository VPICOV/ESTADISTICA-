```python
import matplotlib.pyplot as plt
import numpy as np

# Parámetros
s = 10  # Lado del cuadrado
r = 2   # Radio del círculo

# Crear la figura
plt.figure(figsize=(6, 6))

# Cuadrado
plt.plot([0, s], [0, 0], 'k-')  # Línea inferior
plt.plot([0, s], [s, s], 'k-')  # Línea superior
plt.plot([0, 0], [0, s], 'k-')  # Línea izquierda
plt.plot([s, s], [0, s], 'k-')  # Línea derecha

# Círculo
circle = plt.Circle((s / 2, s / 2), r, color='red', fill=False)
plt.gca().add_artist(circle)

# Añadir los textos
plt.text(s / 2, s * 0.9, f'Área Cuadrado = {s**2}', fontsize=12, ha='center')
plt.text(s / 2, s * 0.1, f'Área Círculo = {round(np.pi * r**2, 2)}', fontsize=12, ha='center', color='red')

# Establecer los límites y etiquetas
plt.xlim(0, s)
plt.ylim(0, s)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Probabilidad Geométrica: Punto en un Área')

# Mostrar el gráfico
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir los límites del cuadrado
xlim = (0, 60)
ylim = (0, 60)

# Crear los datos de la región
x = np.linspace(0, 60, 600)
y = np.linspace(0, 60, 600)
X, Y = np.meshgrid(x, y)

# Calcular la condición de encuentro: |X - Y| <= 15
inside_region = np.abs(X - Y) <= 15

# Graficar la región
plt.figure(figsize=(8, 6))
plt.pcolormesh(X, Y, inside_region, cmap='Blues', shading='auto')

# Agregar las líneas rojas discontinuas
plt.plot(x, x + 15, 'r--', label='$X - Y = 15$')
plt.plot(x, x - 15, 'r--', label='$Y - X = 15$')

# Etiquetas y título
plt.title("Región donde |X - Y| <= 15")
plt.xlabel("Tiempo del primer amigo (minutos)")
plt.ylabel("Tiempo del segundo amigo (minutos)")
plt.xlim(xlim)
plt.ylim(ylim)

# Añadir leyenda
plt.legend()

# Mostrar el gráfico
plt.colorbar(label='Encuentro (Sí/No)')
plt.show()

# Cálculo de la probabilidad
area_total = 60 * 60
area_no_favorable = 2 * 0.5 * 45 * 45
area_favorable = area_total - area_no_favorable
probabilidad = area_favorable / area_total

print(f"La probabilidad de que los dos amigos se encuentren es: {probabilidad:.4f} ({probabilidad * 100:.2f}%)")
```
